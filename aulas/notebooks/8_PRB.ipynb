Para facilitar os cálculos, vamos definir duas funções auxiliares análogas às usadas no notebook de Bayes.

```python
def prob(E):
    """Calcula a probabilidade de um evento booleano E."""
    return E.sum() / E.size  # E.mean()

def prob_cond(A, B):
    """Calcula a probabilidade condicional de A dado B."""
    return prob(A[B])
```

## **Bloco 6. Teorema da Probabilidade Total**

O **Teorema da Probabilidade Total** permite decompor uma probabilidade a partir de uma partição do espaço amostral. Se $A_1, A_2, \ldots, A_k$ formam uma partição, então

$$
P(B) = \sum_{i=1}^{k} P(B \mid A_i)P(A_i)
$$

ou, no caso de dois grupos complementares,

$$
P(B) = P(B \mid A_1)P(A_1) + P(B \mid A_2)P(A_2)
$$
