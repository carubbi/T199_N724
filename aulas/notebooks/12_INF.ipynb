## **Estatística Inferencial e Intervalos de Confiança**

A [estatística inferencial](https://pt.wikipedia.org/wiki/Infer%C3%AAncia_estat%C3%ADstica) compreende métodos que usam resultados de uma amostra para **tomar decisões ou elaborar previsões sobre uma população**. Essa área é fundamental quando o conhecimento sobre a população é limitado ou inexistente — o que ocorre na maioria das aplicações reais. Muitas vezes é caro, demorado ou inviável coletar dados de todos os elementos da população. Assim, coleta‑se uma amostra e calcula‑se uma **estatística amostral apropriada**. A partir do valor dessa estatística, atribui‑se um valor ao parâmetro populacional correspondente (desconhecido).

Esse procedimento, no qual atribuímos um valor numérico a um parâmetro populacional com base nas informações de uma amostra, chama‑se [estimação](https://pt.wikipedia.org/wiki/Estima%C3%A7%C3%A3o). O valor numérico obtido é a estimação do parâmetro. A estatística amostral usada para estimar um parâmetro é chamada de estimador.

## **Estimação Pontual**

Dada uma amostra, o valor da estatística calculada fornece uma estimação pontual do parâmetro populacional correspondente. Por exemplo, a média amostral ($\bar{x}$) é uma estimação pontual da média populacional ($\mu$); o desvio‑padrão amostral ($s$) é uma estimação pontual do desvio‑padrão populacional ($\sigma$).

Como a amostragem é aleatória, amostras diferentes costumam produzir valores diferentes para a estatística amostral (erro amostral). Em outras palavras, $\bar{x}$ e $s$ variam de amostra para amostra, enquanto $\mu$ e $\sigma$ são fixos (mas desconhecidos). Consequentemente, a estimação pontual quase sempre difere do verdadeiro valor populacional. Por isso, é desejável acompanhá‑la de uma medida de precisão — o que motiva o uso de **intervalos de confiança**.

## **Estimação por Intervalo**

Em vez de atribuir um valor único (uma estimação pontual) a um parâmetro populacional, uma **estimação por intervalo** (ou intervalar) fornece uma declaração probabilística, relacionando o intervalo fornecido à probabilidade de que este intervalo realmente contenha o verdadeiro (e desconhecido) parâmetro populacional.

O **nível de confiança** é escolhido *a priori* (isto é, de antemão) e, portanto, depende das preferências do pesquisador. Ele é denotado por $100(1-\alpha)\%$. Embora qualquer valor de nível de confiança possa ser escolhido, os valores mais comuns são 90%, 95% e 99%. Quando expresso como uma probabilidade, o nível de confiança é chamado de **coeficiente de confiança** e é denotado por $(1-\alpha)$. Os coeficientes de confiança mais comuns são 0,90, 0,95 e 0,99, respectivamente.

Um **intervalo de confiança (IC)** de $100(1-\alpha)\%$ é uma estimação intervalar em torno de um parâmetro populacional $\theta$ (aqui, a letra grega $\theta$ (theta) é usada como um parâmetro populacional de interesse, como a média $\mu$ ou o desvio padrão $\sigma$, entre outros) que, sob amostragens aleatórias repetidas de tamanho $N$, espera-se que inclua o verdadeiro valor de $\theta$ em $100(1-\alpha)\%$ das vezes (Lovric 2010).

O valor real que é somado e subtraído da estimação pontual é chamado de **margem de erro**.

$$IC: estimação \ Pontual \pm Margem \ de \ Erro \tag{1}$$

A margem de erro é composta por duas entidades (ou componentes). Primeiro, o chamado **valor crítico** e, segundo, uma medida de variabilidade da **distribuição amostral**.

O **valor crítico** é um valor numérico que corresponde ao nível de confiança definido *a priori*. É denotado como $z^*$. A relação com o nível de confiança é explicitada pelo subscrito $z^*_{\alpha/2}$.

> **Nota:** O intervalo de confiança possui um limite inferior e um superior. Consequentemente, $\alpha$ (o nível de significância, que é $1 - \text{nível de confiança}$) é dividido por 2, pois a área sob a curva além desses limites (nas "caudas" da distribuição) corresponde a $\frac{\alpha}{2} \times 2 = \alpha$.

A medida de variabilidade é o **erro padrão**, denotado como $\frac{\sigma}{\sqrt{n}}$, se $\sigma$ (o desvio padrão populacional) for conhecido. Se $\sigma$ não for conhecido, pode-se usar o erro padrão amostral, dado por $\frac{s}{\sqrt{n}}$, onde $s$ é o desvio padrão da amostra.

Assim, a margem de erro (ME) é expressa como:

$$ME = z^*_{\alpha/2} \times \frac{\sigma}{\sqrt{n}} \tag{2}$$

Vejamos uma figura para melhor compreensão:

<img src="https://drive.google.com/uc?id=1LA7QbFX8xkzbPr86JNuoEK4k-MZ8rvR4" width="800">

Consequentemente, a equação completa para o intervalo de confiança é dada por:

$$IC: estimação \ Pontual \pm z^*_{\alpha/2} \times \frac{\sigma}{\sqrt{n}} \tag{3}$$

Existem duas formas de obter o valor correspondente para $z^*_{\alpha/2}$. Pode-se procurá-lo em uma **tabela de quantis** (*quantile table*) da distribuição Gaussiana padrão (Normal padrão). A outra possibilidade, e a mais moderna (*state-of-the-art*), é utilizar o **Python** para retornar diretamente o valor $z$ correspondente com base no **nível de significância** ($\alpha$) escolhido. Para este propósito, usamos o objeto `norm` e o método `.ppf(<probabilidade>)` do pacote `scipy`.

Vamos construir alguns intervalos de confiança para praticar:

Nível de confiança de 90% ($\alpha = 0.1$)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
lower_90 = stats.norm.ppf(0.05)
upper_90 = stats.norm.ppf(0.95)

print("Os limites inferior e superior do intervalo que cobre uma área de 90% \n",
      "em torno da média são dados por escores-z de",
      round(lower_90, 2), "e", round(upper_90, 2), "respectivamente.")
```

Para um nível de confiança de 90% ($\alpha = 0.1$), a equação (3) acima resulta em:

$$IC_{90\%}: estimação \ Pontual \pm 1.64 \times \frac{\sigma}{\sqrt{n}}$$

Nível de confiança de 95% ($\alpha = 0.05$)

```python
lower_95 = stats.norm.ppf(0.025)
upper_95 = stats.norm.ppf(0.975)

print("The lower and upper limits of the interval that covers an area of 90% \n",
      "around the mean are given by z-scores of",
      round(lower_95, 2), "and", round(upper_95, 2), "respectively.")
```

Para um nível de confiança de 95% ($\alpha = 0.05$), a equação acima resulta em:

$$IC_{95\%}: estimação \ Pontual \pm 1.96 \times \frac{\sigma}{\sqrt{n}}$$

Nível de confiança de 99% ($\alpha = 0.01$)

Para um nível de confiança de 99% ($\alpha = 0.01$), a equação acima resulta em:

$$IC_{99\%}: estimação \ Pontual \pm 2.58 \times \frac{\sigma}{\sqrt{n}}$$

## **A Largura de um Intervalo de Confiança e o Trade-Off entre Precisão e Acurácia**

### Um experimento mental

Considere a temperatura média diária em **Fortaleza** em um dia típico. Agora, a tarefa é **estimar a temperatura média de amanhã**. Digamos que medimos uma temperatura média diária de **27°C**.

Existem várias abordagens, mais ou menos sofisticadas, para resolver essa tarefa. Pode-se pesquisar dados arquivados e basear a estimação em estatísticas descritivas, ou pode-se até aplicar uma abordagem de modelagem muito sofisticada com inúmeros parâmetros, ou pode-se simplesmente fazer um palpite com base no senso comum. No entanto, não importa a abordagem escolhida, o resultado será sempre uma **estimação, associada a algum nível de incerteza**.

Para refletir essa incerteza, não apresentamos uma estimação pontual da temperatura de amanhã, mas sim uma estimação intervalar. A fim de alcançar uma alta **acurácia** (*confiabilidade, certeza ou validade*) em nossa estimação, o que significa que queremos estar muito confiantes de que nosso intervalo contém o valor real, aplicamos uma margem de erro muito grande. Por exemplo, afirmamos que a temperatura de amanhã é **$27 \pm 10 \text{ °C}$**. Com base no senso comum, provavelmente concordaríamos que é muito provável que a temperatura média diária de qualquer dia em Fortaleza esteja entre **17°C e 37°C**.

Embora o intervalo seja muito grande e possa até conter a maioria das temperaturas médias diárias do histórico de observação meteorológica de Fortaleza (cobrindo desde madrugadas excepcionalmente "frias" até picos de calor extremos), ainda há uma pequena chance de estarmos errados. Imagine qualquer evento cataclísmico natural ou causado pelo homem, como um tsunami no litoral, o impacto de um asteroide ou uma guerra nuclear; nesses casos felizmente muito improváveis, mesmo uma margem de erro tão grande pode não garantir que a temperatura média de amanhã esteja dentro do intervalo fornecido. No entanto, é essencial notar que, **para alcançar alta acurácia, aumentamos o nível de confiança e, assim, a largura do intervalo de confiança.**

Afirmamos que a temperatura de amanhã está na faixa de $27 \pm 10 \text{ °C}$. No entanto, tal afirmação, apesar da alta acurácia, tem algum valor? Essa estimação nos ajuda a decidir que roupa vestir amanhã (afinal, há uma diferença enorme entre 17°C e 37°C)? Não, de forma alguma!

Assim, em muitas aplicações, não estamos interessados apenas na **acurácia**, mas também na **precisão** de uma estimação. Um palpite da temperatura de amanhã com maior precisão seria **$27 \pm 1 \text{ °C}$** (ou seja, entre 26°C e 28°C). Tal previsão nos ajuda a decidir o que esperar (será um dia quente dentro da normalidade); no entanto, a chance de estarmos errados (a média real ser 25,5°C ou 28,5°C) é muito maior.

> **Nota:** É importante memorizar que, ao aumentar a **precisão**, diminuímos a largura do intervalo de confiança e, consequentemente, diminuímos o **nível de confiança** (que está associado à acurácia). Esse *trade-off* (relação de troca) entre acurácia e precisão ao selecionar um nível de confiança adequado é muito importante em aplicações da vida real.

## **Estimação da Média Populacional - A Distribuição z**

Estimar a média populacional a partir de uma amostra aleatória é uma tarefa comum. Suponha que o desvio padrão da população ($\sigma$) seja conhecido. Nesse caso, a construção de um intervalo de confiança para a média da população ($\mu$) é baseada na distribuição amostral das médias amostrais, que segue uma distribuição normal (garantida pelo **Teorema do Limite Central**). Lembre-se que, se a população da qual a amostra é retirada não for normalmente distribuída, o tamanho da amostra ($n$) deve ser $ > 30$.

O intervalo de confiança de $100(1-\alpha)\%$ para $\mu$ é dado por:

$$IC: \bar{x} \pm z^*_{\alpha/2} \times \sigma_{\bar{x}}$$

onde $\sigma_{\bar{x}} = \frac{\sigma}{\sqrt{n}}$

O valor de $z^*_{\alpha/2}$ corresponde ao valor crítico e é obtido da **tabela normal padrão** ou calculado diretamente em Python usando a função `norm.ppf()` do pacote `scipy`. O valor crítico é uma quantidade relacionada ao nível de confiança desejado. Em outras palavras, ele é multiplicado pelo erro padrão, dado por $\sigma_{\bar{x}}$, para alargar ou estreitar a margem de erro. Valores típicos para $z^*_{\alpha/2}$ são 1.64, 1.96 e 2.58, correspondendo a níveis de confiança de 90%, 95% e 99%, respectivamente.

O erro padrão ($\sigma_{\bar{x}}$) é dado pela razão entre o desvio padrão da população ($\sigma$) e a raiz quadrada do tamanho da amostra ($n$). Ele descreve o grau em que a estatística amostral calculada pode ser esperada para diferir de uma amostra para outra. O produto do valor crítico e do erro padrão é chamado de **margem de erro**. Essa quantidade é subtraída e adicionada ao valor de $\bar{x}$ para obter o intervalo de confiança para $\mu$.

```python
# Cria a figura e o eixo do plot
fig, ax = plt.subplots(figsize=(10, 6))

# Plota o histograma
# `density=True` normaliza o histograma para que a área sob as barras some 1,
ax.hist(females["height"],
        bins=29,
        color="gainsboro",
        edgecolor="black",
        density=True,
        label='Histograma de Alturas')

# Gera valores para o eixo x para a função PDF
x = np.linspace(np.min(females["height"]),
                np.max(females["height"]), 1000)

# Calcula os valores da Função Densidade de Probabilidade (PDF) para a distribuição normal
heights_pdf = stats.norm.pdf(x,
                       loc=mean_heights,
                       scale=sigma_heights)

# Plota a PDF sobre o histograma
ax.plot(x,
        heights_pdf,
        color="red",
        label=f'PDF Normal (μ={mean_heights:.2f}, σ={sigma_heights:.2f})')

# Adiciona título e rótulos aos eixos
ax.set_title('Histograma da distribuição de altura feminina com PDF Normal sobreposta')
ax.set_xlabel('Altura (cm)')
ax.set_ylabel('Densidade de Probabilidade')
ax.legend() # Mostra a legenda
ax.grid(axis='y', alpha=0.75) # Adiciona um grid (opcional)

plt.show()
```

```python
np.random.seed(40)
sample = np.random.normal(loc = mean_heights,
                          scale = sigma_heights,
                          size = 10)
mean_sample = np.mean(sample)

print(mean_sample)
```

Nossa amostra aleatória produz uma média amostral ($\bar{x}$) de aproximadamente:

Esta é a nossa estimação pontual para o parâmetro populacional de interesse, que neste caso é a altura média das estudantes do sexo feminino ($\mu$).

Quão precisa é a nossa estimação pontual? Podemos verificar se nossa estimação é igual ao verdadeiro parâmetro populacional:

```python
mean_sample == mean_heights
```

*OK, já esperávamos por isso.*

Vamos calcular algumas estimações intervalares construindo os intervalos de confiança de 90%, 95% e 99%. Relembre a equação para um intervalo de confiança:

$$IC: estimação \ Pontual \pm z^{*}_{\alpha/2} \times \frac {\sigma} {\sqrt {n}}$$

O valor crítico $z^{*}_{\alpha/2}$ resulta em 1.64, 1.96 e 2.58 para níveis de confiança de 90%, 95% e 99%, respectivamente.

Aplicando aos nossos dados de altura, a equação produz o seguinte:

$$IC_{90\%} : 165.04 \pm 1.64 \times \frac {7.92} {\sqrt {10}} = 165.04 \pm 4.12$$
Podemos afirmar que temos 90% de confiança de que a altura média das estudantes do sexo feminino (o parâmetro populacional $\mu$) está entre 160.92 e 169.16 cm.

$$IC_{95\%} : 165.04 \pm 1.96 \times \frac {7.92} {\sqrt {10}} = 165.04 \pm 4.91$$
Podemos afirmar que temos 95% de confiança de que a altura média das estudantes do sexo feminino está entre 160.13 e 169.95 cm.

$$IC_{99\%} : 165.04 \pm 2.58 \times \frac {7.92} {\sqrt {10}} = 165.04 \pm 6.45$$
Podemos afirmar que temos 99% de confiança de que a altura média das estudantes do sexo feminino está entre 158.59 e 171.49 cm.

> **Nota:** É óbvio que, se quisermos ter uma **confiança maior** de que o parâmetro populacional desconhecido está contido no intervalo, a **margem de erro se torna maior**.

Para validar nosso resultado, vamos investigar se capturamos o verdadeiro valor populacional dentro de nossas estimações por intervalo. É importante lembrar que o intervalo de confiança não atribui uma probabilidade à nossa média amostral. No entanto, ele afirma que, sob amostragem aleatória repetida, o intervalo de confiança é esperado para incluir a média populacional em $100(1-\alpha)\%$ das vezes. Para testar essa afirmação, vamos escrever uma função.

```python
def CI_evaluation(pop_mean, sigma, n, estimate, alpha_levels):
    """
    Avalia se a média populacional verdadeira está contida em intervalos de confiança
    para diferentes níveis de significância.

    Args:
        pop_mean (float): A média verdadeira da população.
        sigma (float): O desvio padrão verdadeiro da população.
        n (int): O tamanho da amostra usada para a estimativa.
        estimate (float): A média amostral (estimativa pontual).
        alpha_levels (list): Uma lista de níveis de significância (e.g., [0.10, 0.05, 0.01]).

    Returns:
        list: Uma lista de booleanos (True/False) indicando se a média populacional
              foi capturada por cada intervalo de confiança correspondente ao alpha_level.
    """
    # Lista para armazenar os resultados (True/False) para cada nível alpha
    check_results = []

    for alpha_value in alpha_levels:
        # Calcula o valor crítico z*
        critical_value_z = stats.norm.ppf(1 - (alpha_value / 2))

        # Calcula o Erro Padrão
        standard_error = sigma / np.sqrt(n)

        # Calcula a Margem de Erro
        margin_of_error = critical_value_z * standard_error

        # Calcula os Limites Inferior e Superior do Intervalo de Confiança (IC)
        lower_bound = estimate - margin_of_error
        upper_bound = estimate + margin_of_error

        # Verifica se a Média Populacional (pop_mean) está dentro do Intervalo de Confiança
        is_captured = (pop_mean >= lower_bound) and (pop_mean <= upper_bound)

        check_results.append(is_captured)

    return check_results
```

```python
checks = CI_evaluation(mean_heights, sigma_heights, 10, mean_sample, [0.10, 0.05, 0.01])
print(checks)
```

```python
df = pd.DataFrame(checks, columns = ["evaluation"])
df.index = ['90 %', '95 %', '99 %']
df
```

Bem, um resultado interessante. Todos os três intervalos de confiança capturam a verdadeira média populacional (`mean_heights` = 163.65). Obviamente, a média de nossa amostra aleatória (`mean_sample` = 165.04) foi um estimador bastante bom.

Até agora, focamos apenas em uma amostra aleatória particular. No entanto, o que devemos esperar se repetirmos o processo de amostragem várias e várias vezes?

Um intervalo de confiança de qualquer nível de confiança particular afirma que, sob amostras aleatórias repetidas, o intervalo de confiança é esperado para incluir o verdadeiro parâmetro populacional em $100(1-\alpha)\%$ das vezes. Assim, um intervalo de confiança de 99% implica que, em média, em 99 de 100 casos, a margem de erro é grande o suficiente para abranger o verdadeiro valor populacional. Isso também implica que estamos errados em média uma vez em 100. Além disso, para um intervalo de confiança de 95% ou 90%, estamos errados 5 ou 10 vezes em 100, respectivamente. Usaremos os recursos do **Python** para testar essa afirmação, conduzindo um experimento de simulação.

Realizaremos uma amostragem da distribuição de `heights` (alturas) que vimos acima. Tiraremos 10, 50, 100, 1000 e 10000 amostras aleatórias. Cada amostra terá um tamanho de $n=10$. Calcularemos a média amostral $\bar{x}$, que é nossa estatística amostral de interesse para cada amostra. Aplicando nossa função `CI_evaluation`, testaremos se a verdadeira média populacional é capturada pelos intervalos de confiança de 90%, 95% e 99%. Armazenaremos o vetor booleano resultante e calcularemos a proporção de vezes que cada intervalo de confiança particular realmente contém a verdadeira média populacional. Finalmente, armazenaremos a proporção como porcentagens em um objeto `DataFrame` do `pandas` e renomearemos suas linhas e colunas para melhor legibilidade.

```python
trials = [10, 50, 100, 1000, 10000]
sigs = ['90', '95', '99']
results = []

for trial in trials:
    proofs = []
    for i in range(1, trial + 1):
        # Gera uma amostra aleatória da distribuição normal
        sample = np.random.normal(loc = mean_heights,
                                  scale = sigma_heights,
                                  size = 10)

        mean_sample = np.mean(sample)

        # Avalia se os intervalos de confiança capturam a média populacional verdadeira
        check = CI_evaluation(mean_heights, sigma_heights, 10, mean_sample, [0.10, 0.05, 0.01])
        proofs.append(check)

    # Converte a lista de resultados booleanos em um DataFrame
    proofs = pd.DataFrame(proofs, columns = sigs)

    res = []
    for col in sigs:
        # Conta quantas vezes o intervalo de confiança capturou a média populacional
        # E calcula a proporção (probabilidade empírica)
        res.append(proofs.loc[:,col].value_counts().get(True, 0) / trial)

    results.append(res)
```

```python
df = pd.DataFrame(results, columns = sigs)
df.index = trials
df
```

## **Estimação da Média Populacional - A Distribuição t**

Até agora, dependemos de $\sigma$, o desvio padrão populacional, para inferir a média populacional. O parâmetro populacional $\sigma$ é usado para calcular o erro padrão ($SE = \frac{\sigma}{\sqrt{n}}$), que é um dos constituintes da margem de erro. No entanto, e se não soubermos o desvio padrão populacional, como geralmente é o caso? Pode-se usar o desvio padrão amostral, denotado por $s$, como um estimador para o desvio padrão populacional.

Se $s \approx \sigma$, então $SE = \frac{s}{\sqrt{n}}$

Devemos notar que, em contraste com $\sigma$, o desvio padrão amostral $s$ variará de amostra para amostra e que $s < \sigma$. Ao aumentar o tamanho da amostra $n$, $s$ se tornará uma estimação melhor para $\sigma$. No entanto, não importa o que aconteça, enquanto não soubermos $\sigma$, teremos que estimar duas quantidades ao realizar o procedimento de inferência: a média $\mu$ e o desvio padrão $\sigma$. É por isso que usar $s$ como estimador para $\sigma$ tende a adicionar mais incerteza à estimação da média $\mu$. Para contrariar essa incerteza adicional, aplicamos a chamada **distribuição t** ou **Distribuição t de Student** para calcular a margem de erro ($ME$).

A distribuição t de Student recebeu o nome em homenagem a **William Sealy Gosset** (1876-1937), que a determinou pela primeira vez em 1908. Gosset foi um dos melhores graduados em química e matemática de Oxford em sua geração. Em 1899, ele aceitou um emprego como cervejeiro na **Arthur Guinness Son & Co, Ltd** em Dublin, Irlanda. Trabalhando para a cervejaria Guinness, ele estava interessado no controle de qualidade baseado em pequenas amostras em várias etapas do processo de produção. Como a Guinness proibia seus funcionários de publicar artigos para evitar a divulgação de informações confidenciais, Gosset publicou seu trabalho sob o pseudônimo de "Student", e sua identidade não foi conhecida por algum tempo após a publicação de suas maiores conquistas, então a distribuição foi chamada de distribuição de Student ou **distribuição t**, deixando seu nome menos conhecido do que seus importantes resultados em estatística (Lovric 2011).

A curva da distribuição t é, assim como a curva da distribuição normal, simétrica (em forma de sino) em torno da média. No entanto, a curva da distribuição t é mais achatada do que a curva da distribuição normal padrão. Consequentemente, a curva da distribuição t tem uma altura menor e uma dispersão maior do que a distribuição normal padrão.

A distribuição t possui apenas um parâmetro, chamado **graus de liberdade** ($df$). A forma de uma curva de distribuição t particular depende do número de graus de liberdade ($df$). O número de graus de liberdade para uma distribuição t é igual ao tamanho da amostra menos um, ou seja, $df = n-1$.

À medida que o tamanho da amostra, $n$, e consequentemente $df$, aumenta, a distribuição t se aproxima da distribuição normal padrão. As unidades de uma distribuição t são denotadas por $t$. A média da distribuição t é igual a $0$, e seu desvio padrão é $\sqrt{df/(df-2)}$ ([Mann 2012](https://archive.org/details/IntroductoryStatistics7thEdP.MannWiley2010WW)).

```python
x = np.arange(-4, 4, 0.1)
df = [1, 2, 8, 30] # Graus de liberdade
color = ["red", "blue", "darkgreen", "gold"]
handle = ["df=1", "df=3", "df=8", "df=30"]

plt.figure(figsize=(10, 5))
plt.title("Comparação de Distribuições t", fontsize=16)

# Adiciona a distribuição normal padrão para comparação
plt.plot(
    x, stats.norm.pdf(x, 0, 1), color="black", linestyle="--", label="Normal Padrão"
)

# Plota as diferentes distribuições t
for d, c, h in zip(df, color, handle):
    y = stats.t.pdf(x, df=d, loc=0, scale=1) # loc é a média, scale é o desvio padrão
    plt.plot(x, y, color=c, label=h)

# Adiciona uma linha vertical no centro (média)
plt.axvline(
    x=0,
    color="black",
    linestyle="dashed",
)

# Adiciona o texto para indicar a média
plt.text(-0.2, -0.05, ("$\\mu = 0$"), fontsize=14)

# Adiciona a legenda
plt.legend(fontsize=14)

# Remove os eixos para uma visualização mais limpa (estilo de slide)
plt.axis("off")

# Exibe o gráfico
plt.show()
```

Propriedades Básicas das Curvas t ([Weiss, 2010](https://www.sciencedirect.com/book/9780128043172/introductory-statistics))

* A área total sob uma curva t é igual a 1.
* Uma curva t se estende indefinidamente em ambas as direções, aproximando-se, mas nunca tocando, o eixo horizontal.
* Uma curva t é simétrica em torno de 0.
* À medida que o número de graus de liberdade aumenta, as curvas t se parecem cada vez mais com a curva normal padrão.

O procedimento para obter um intervalo de confiança para uma média populacional quando o desvio padrão populacional $\sigma$ é desconhecido é essencialmente o mesmo de quando o desvio padrão populacional é conhecido. A distribuição t e o desvio padrão amostral $s$ são invocados em vez da distribuição normal padrão (escores z) e do desvio padrão populacional $\sigma$.

Relembre a construção de um intervalo de confiança:

$$IC : \text{estimação pontual} \pm ME$$

A margem de erro ($ME$) contém o valor crítico e uma medida para a variabilidade da distribuição amostral. O valor crítico é $t^*_{df, \alpha/2}$ para o nível de confiança e graus de liberdade dados. Seu valor é obtido de uma **tabela de distribuição t** para $n-1$ graus de liberdade, ou é calculado diretamente em Python usando a função `t.ppf()` do pacote `scipy`. A medida para a variabilidade da distribuição amostral é o **erro padrão** $SE$. Como o desvio padrão populacional $\sigma$ não é conhecido, ele é substituído pelo desvio padrão amostral $s$, resultando, assim, em $SE = \frac{s}{\sqrt{n}}$.

Consequentemente, o intervalo de confiança de $100(1-\alpha)\%$ para $\mu$ é dado por:

$$IC : \bar{x} \pm t^*_{df, \alpha/2} \times \frac{s}{\sqrt{n}}$$

Vamos construir alguns intervalos de confiança para praticar! Para este exercício, $df$ (graus de liberdade) é definido como 12.

Nível de confiança de 90% ($\alpha = 0.1$)

```python
from scipy.stats import t

lower_90 = t.ppf(0.05, df = 12)
upper_90 = t.ppf(0.95, df = 12)

print("Os limites inferior e superior do intervalo que cobre uma área de 90% \n",
      "em torno da média são dados por valores t (gl=12) de",
      round(lower_90, 2), "e", round(upper_90, 2), ", respectivamente.")
```

Portanto, para um nível de confiança de 90% ($\alpha = 0.1$), a equação acima resulta em:

$$IC_{90\%} : \text{estimação Pontual} \pm 1.78 \times \frac{s}{\sqrt{n}}$$

Nível de confiança de 95% ($\alpha = 0.05$)

```python
alpha = 0.05
lower_95 = t.ppf(alpha / 2, df = 12)
upper_95 = t.ppf(1 - (alpha / 2) , df = 12)

print("Os limites inferior e superior do intervalo que cobre uma área de\n",
      (1 - alpha) * 100, # Multiplicar por 100 para mostrar como porcentagem
      "% em torno da média são dados por valores t (gl=12) de",
      round(lower_95, 2), "e", round(upper_95, 2), ", respectivamente.")
```

Portanto, para um nível de confiança de 95% ($\alpha = 0.05$), a equação acima resulta em:

$$IC_{95\%} : \text{estimação Pontual} \pm 2.18 \times \frac{s}{\sqrt{n}}$$

```python
alpha = 0.01
lower_99 = t.ppf(alpha / 2, df = 12)
upper_99 = t.ppf(1 - (alpha / 2) , df = 12)

print("Os limites inferior e superior do intervalo que cobre uma área de\n",
    (1 - alpha) * 100, # Multiplicar por 100 para mostrar como porcentagem
    "% em torno da média são dados por valores t (gl=12) de/n",
    round(lower_99, 2), "e", round(upper_99, 2), ", respectivamente.")
```

$$IC_{99\%} : \text{estimação Pontual} \pm 3.05 \times \frac{s}{\sqrt{n}}$$

## **Resumo Geral**

### **Quando usar a Distribuição Z (Normal Padrão)?**

Você deve usar a distribuição Z quando:

1.  **O desvio padrão da população ($\sigma$) é CONHECIDO.**
    * Este é o cenário ideal, mas raramente acontece na prática. Geralmente, se você conhece $\sigma$, é porque já conhece a população muito bem, e a estimação por IC talvez nem fosse necessária.
    * Nesse caso, a fórmula para o IC da média é:
        $$IC: \bar{x} \pm z^*_{\alpha/2} \times \frac{\sigma}{\sqrt{n}}$$

2.  **O tamanho da amostra ($n$) é GRANDE ($n \ge 30$), e o desvio padrão da população ($\sigma$) é DESCONHECIDO.**
    * Neste cenário, podemos usar o desvio padrão da amostra ($s$) como uma boa estimativa para $\sigma$. Graças ao **Teorema do Limite Central**, para amostras grandes, a distribuição amostral das médias se aproxima da normal, mesmo que a população original não seja normal.
    * Portanto, a fórmula se aproxima da Z:
        $$IC: \bar{x} \pm z^*_{\alpha/2} \times \frac{s}{\sqrt{n}}$$
    * **Observação:** Embora tecnicamente a distribuição t ainda seja mais precisa aqui, para $n \ge 30$, os valores críticos $z^\*$ e $t^\*$ são muito próximos, e muitos livros e softwares ainda permitem o uso da Z como uma boa aproximação.

### **Quando usar a Distribuição t de Student?**

Você deve usar a distribuição t de Student quando:

1.  **O desvio padrão da população ($\sigma$) é DESCONHECIDO, E o tamanho da amostra ($n$) é PEQUENO ($n < 30$).**
    * Este é o cenário mais comum na prática. Você não sabe $\sigma$ e sua amostra não é grande o suficiente para a aproximação Z ser totalmente robusta.
    * Nesse caso, você usa o desvio padrão da amostra ($s$) no cálculo do erro padrão, e o valor crítico é obtido da distribuição t, que leva em consideração a incerteza adicional de estimar $\sigma$ com $s$.
    * A fórmula para o IC da média é:
        $$IC: \bar{x} \pm t^*_{df, \alpha/2} \times \frac{s}{\sqrt{n}}$$
    * Onde $df$ (graus de liberdade) = $n-1$.
    * **Condição Adicional:** A população de onde a amostra foi retirada deve ser **aproximadamente normal**. Se a população não for normal e a amostra for pequena, nem a distribuição Z nem a t são apropriadas.

### **Fluxograma Mental:**

1.  **$\sigma$ (Desvio Padrão Populacional) Conhecido?**
    * **SIM:** Use **Z-distribuição**.
    * **NÃO:** Vá para a próxima pergunta.

2.  **$n$ (Tamanho da Amostra) $\ge 30$?**
    * **SIM:** Use **Z-distribuição** (ou t-distribuição, pois são muito próximas aqui). A Z é aceitável devido ao Teorema do Limite Central.
    * **NÃO ($n < 30$):** Use **t-distribuição**, mas **verifique se a população é aproximadamente normal**.

Em ambientes de software modernos (como Python com `scipy` ou R), a distinção entre $n \ge 30$ e $n < 30$ torna-se menos crítica para a *escolha da distribuição* se $\sigma$ é desconhecido. A prática mais rigorosa é **sempre usar a distribuição t** quando $\sigma$ é desconhecido, independentemente do tamanho da amostra, pois a t converge para a Z à medida que $n$ (e $df$) aumenta, então ela é a escolha mais conservadora e precisa. No entanto, o critério de $n \ge 30$ para usar Z (aproximadamente) ainda é amplamente ensinado e útil para entender o Teorema do Limite Central.
