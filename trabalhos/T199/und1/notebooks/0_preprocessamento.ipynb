### **Validação e Transformações**




**Objetivo da seção**

#### **Validação estrutural**


```python
# Define colunas mínimas para o pipeline.
required_cols = [
    'year', 'name', 'playerId', 'teamId', 'positionId', 'status',
    'salary', 'birthdate', 'height', 'weight', 'experience'
]
print(f'Colunas: {required_cols}')
```

```python
# Define chave de unicidade por temporada/jogador/time.
key_cols = ['year', 'playerId', 'teamId']

# Conta duplicidade pela chave de unicidade.
dup_key = df.duplicated(subset=key_cols).sum()

# Conta duplicidade de linha completa.
dup_full = df.duplicated().sum()

# Define quantas colunas faltantes principais exibir.
top_n = 15

# Reporta duplicidades e faltantes principais.
print('Duplicidades (linha completa):', dup_full)
print(f'Duplicidades por chave {key_cols}:', dup_key)
print(f'Faltantes (top {top_n}):')
display(df.isna().sum().sort_values(ascending=False).head(top_n))
```

#### **Limpeza básica de textos e normalização de faltantes**

```python
# Identifica colunas textuais para tratamento padronizado.
str_cols = df.select_dtypes(include=['object', 'str']).columns.tolist()

# Remove espaços extras nas extremidades.
for col in str_cols:
    df[col] = df[col].astype('string').str.strip()

# Define marcadores textuais de ausência.
missing_tokens = {'', 'NA', 'N/A', 'null', 'None', '-'}

# Converte marcadores para NA do pandas.
for col in str_cols:
    df[col] = df[col].replace(list(missing_tokens), pd.NA)

# Informa quantidade de colunas textuais tratadas.
print('Colunas textuais tratadas:', len(str_cols))
```

#### **Correção de inconsistência**

```python
# Detecta casos em que altura veio em libras.
height_has_lbs = df['height'].astype('string').str.contains(
    r'lbs', case=False, na=False
)
print('Registros com "lbs" em height:', height_has_lbs.sum())
print(df['height'].dropna().head())
```

```python
# Detecta casos em que peso veio em pés/polegadas.
weight_has_feet = df['weight'].astype('string').str.contains(
    r"\d+'\s*\d+\"", na=False
)
print('Registros com formato feet/inches em weight:', weight_has_feet.sum())
print(df['weight'].dropna().head())
```

```python
# Marca linhas com possível inversão entre altura e peso.
swap_mask = height_has_lbs & weight_has_feet

# Quantifica linhas suspeitas.
swap_count = int(swap_mask.sum())
if swap_count > 0:
    # Guarda temporariamente os valores atuais.
    h_old = df.loc[swap_mask, 'height'].copy()
    w_old = df.loc[swap_mask, 'weight'].copy()
    # Executa troca de valores entre colunas.
    df.loc[swap_mask, 'height'] = w_old
    df.loc[swap_mask, 'weight'] = h_old

# Reporta o total de correções aplicadas.
print('Linhas corrigidas por inversão height/weight:', swap_count)
```

#### **Conversões tipadas**

Conversões de tipo sem criação de novas variáveis analíticas.

**Métodos usados nesta subseção**
- `pd.to_datetime(..., errors='coerce')`: converte datas e preserva inválidos como ausentes.
- `pd.to_numeric(..., errors='coerce')`: converte números textuais e protege o fluxo contra valores inválidos.
- `.astype('Int64')`: define inteiro nullable para colunas discretas com possíveis ausências.


```python
# Converte salário textual para numérico em dólar.
df['salary_usd'] = (
    df['salary']
    .str.replace('$', '', regex=False)
    .str.replace(',', '', regex=False)
    .astype('Float64')
)
print('Salário convertido para numérico:', df['salary_usd'].notna().sum())
print(df[['salary', 'salary_usd']].head())
```

```python
# Converte data de nascimento para datetime.
df['birthdate_dt'] = pd.to_datetime(df['birthdate'], errors='coerce')

print('Birthdate convertido para datetime:', df['birthdate_dt'].notna().sum())
print(df[['birthdate', 'birthdate_dt']].head())
```

```python
# Converte ano da temporada para tipo numérico inteiro nullable.
df['year'] = pd.to_numeric(df['year'], errors='coerce').astype('Int64')

print('Ano convertido para Int64:', df['year'].notna().sum())
print(df['year'].head())
```

#### **Engenharia de variáveis**

Criação de variáveis derivadas a partir de colunas já tipadas.

**Métodos usados nesta subseção**
- Operações vetorizadas com séries pandas para derivação de variáveis.
- `str.extract(...)` para extração de componentes numéricos via regex.
- `.astype(...)` para tipagem final consistente das variáveis derivadas.


```python
# Calcula idade na temporada a partir do ano já tipado e da data de nascimento.
df['age_in_season'] = (
    df['year'] - df['birthdate_dt'].dt.year
).astype('Int64')

print('Idade na temporada calculada:', int(df['age_in_season'].notna().sum()))
df['age_in_season'].head()
```

```python
# Extrai número de anos de experiência da string original.
df['experience_years'] = (
    df['experience']
    .str.extract(r'(\d+)', expand=False)
    .astype('Int64')
)

print('Anos de experiência calculados:', int(df['experience_years'].notna().sum()))
df[['experience', 'experience_years']].dropna().head()
```

**Explicação da regex `r'(\d+)'` (experiência)**

- `r"..."`: string *raw* no Python.
- `( ... )`: grupo de captura (retorna o trecho encontrado).
- `\d+`: um ou mais dígitos consecutivos.

Na prática, extrai o número de textos como `"3 years"`, `"10"` ou `"2nd"`.
Com `expand=False`, o retorno é uma `Series` (não `DataFrame`).


```python
# Extrai pés/polegadas da altura textual e converte para metros.
ht = df['height'].str.extract(
    r"(?P<ft>\d+)'\s*(?P<inch>\d+)"
)
print('Alturas com formato feet/inches extraídas:', ht.notna().all(axis=1).sum())
print(ht.dropna().head())
```

**Explicação da regex usada na extração de altura**

- `r"..."`: string *raw* no Python (evita escapes extras).
- `(?P<ft>\d+)`: grupo nomeado `ft` com 1+ dígitos (pés).
- `'`: apóstrofo literal entre pés e polegadas.
- `\s*`: zero ou mais espaços entre os blocos.
- `(?P<inch>\d+)`: grupo nomeado `inch` com 1+ dígitos (polegadas).

Exemplo: `6' 7"` resulta em `ft = 6` e `inch = 7`.


```python
# Converte para metros usando fórmula: altura_m = (pés * 0.3048) + (polegadas * 0.0254)
df['height_m'] = (
    ht['ft'].astype('Float64') * 0.3048 +
    ht['inch'].astype('Float64') * 0.0254
)
print('Altura convertida para metros:', df['height_m'].notna().sum())
print(df[['height', 'height_m']].dropna().head())
```

```python
# Extrai peso em libras e converte para quilogramas.
weight_raw = pd.to_numeric(
    df['weight'].str.extract(r'(\d+)', expand=False),
    errors='coerce'
)
print('Pesos com formato numérico extraídos:', weight_raw.notna().sum())
print(weight_raw.dropna().head())
```

```python
# Converte pesos em libras para kg.
df['weight_kg'] = (weight_raw * 0.45359237).astype('Float64')

print('Peso em kg calculado:', int(df['weight_kg'].notna().sum()))
df[['weight', 'weight_kg']].dropna().head()
```

```python
# Preenche ausência de posição com rótulo explícito.
df['positionId'] = df['positionId'].fillna('NA_POS')

print('positionId faltante após preenchimento:', int(df['positionId'].isna().sum()))
df[['positionId']].head()
```

```python
# Lista de colunas para inspeção consolidada das variáveis derivadas.
cols = [
    'year', 'salary', 'salary_usd', 'birthdate', 'birthdate_dt',
    'age_in_season', 'height', 'height_m', 'weight', 'weight_kg',
    'experience', 'experience_years'
]

df[cols].head()
```

```python
# Exibe tipos e não nulos das colunas convertidas/derivadas.
df[cols].info()
```

### **Dados Plausíveis**


**Bibliotecas, funções e métodos usados nesta subseção**
- `pandas`: aplicação de filtros vetorizados sobre variáveis derivadas.
- Operadores lógicos (`<`, `>`, `|`): definem limites de plausibilidade.
- `fillna(False)` e `sum()`: evitam contagem indevida e totalizam fora da faixa.


```python
# 5.4) Faixa plausível de idade
# Conta idades fora do intervalo esperado para atletas da NBA.
age_out = int(
    ((df['age_in_season'] < 15) | (df['age_in_season'] > 50))
    .fillna(False).sum()
)
print('age_out_of_range_<15_or_>50:', age_out)
```

```python
# 5.6) Faixa plausível de peso (kg)
# Conta pesos fora do intervalo plausível definido.
weight_out = int(
    ((df['weight_kg'] < 63) | (df['weight_kg'] > 155))
    .fillna(False).sum()
)
print('weight_out_of_range_<63_or_>155:', weight_out)
```

### **Seleção de colunas para exportação**

```python
# Remove campos brutos que já possuem versão tratada.
cols_drop_export = [
    'salary',
    'birthdate',
    'height',
    'weight',
    'experience',
]

# Remove todas as colunas de média por regra do projeto.
avg_cols = [c for c in df.columns if c.startswith('avg')]
cols_drop_export = cols_drop_export + avg_cols

print('Colunas a remover na exportação:', cols_drop_export)
```

```python
# Gera base de exportação com colunas filtradas.
df_export = df.drop(columns=cols_drop_export, errors='ignore').copy()

# Remove colunas sem dado (100% ausentes).
empty_cols = [c for c in df_export.columns if df_export[c].isna().all()]
df_export = df_export.drop(columns=empty_cols, errors='ignore')

# Exibe as primeiras linhas da base final para validação visual.
df_export.head(3)    
```

```python
# Lista colunas removidas na regra fixa + avg.
removed_cols = [c for c in cols_drop_export if c in df.columns]

# Reporta dimensões e colunas removidas.
print('Shape original:', df.shape)
print('Shape exportação:', df_export.shape)
print('Qtde colunas removidas (fixas + avg):', len(removed_cols))
print('Colunas removidas (fixas + avg):')
for col in removed_cols:
    print('-', col)
```

```python
# 6.1) Dicionário EN -> PT-BR (termos técnicos NBA)
# Define mapeamento de nomes para padrão PT-BR da disciplina.
rename_cols_pt_nba = {
    'year': 'temporada',
    'name': 'nome_jogador',
    'playerId': 'id_jogador',
    'playerSlug': 'id_url_jogador',
    'positionId': 'sigla_posicao',
    'teamId': 'sigla_time',
    'status': 'status_jogador',
    'gamesPlayed': 'jogos_disputados',
    'flagrantFouls': 'faltas_flagrantes',
    'technicalFouls': 'faltas_tecnicas',
    'ejections': 'expulsoes',
    'doubleDouble': 'duplo_duplo',
    'tripleDouble': 'triplo_duplo',
    'minutes': 'minutos_totais',
    'rebounds': 'rebotes',
    'fouls': 'faltas',
    'fieldGoalPct': 'pct_arremessos_quadra',
    'threePointFieldGoalPct': 'pct_bolas_3',
    'freeThrowPct': 'pct_lances_livres',
    'points': 'pontos',
    'fieldGoalsMade': 'arremessos_quadra_convertidos',
    'fieldGoalsAttempted': 'arremessos_quadra_tentados',
    'threePointFieldGoalsMade': 'bolas_3_convertidas',
    'threePointFieldGoalsAttempted': 'bolas_3_tentadas',
    'freeThrowsMade': 'lances_livres_convertidos',
    'freeThrowsAttempted': 'lances_livres_tentados',
    'assists': 'assistencias',
    'turnovers': 'perdas_bola',
    'steals': 'roubos_bola',
    'blocks': 'tocos',
    'position': 'posicao',
    'college': 'universidade',
    'draftinfo': 'informacoes_recrutamento',
    'birthplace': 'local_nascimento',
    'team': 'nome_time',
    'rankingSalary': 'ranking_salarial',
    'salary_usd': 'salario_usd',
    'birthdate_dt': 'data_nascimento',
    'age_in_season': 'idade_temporada',
    'height_m': 'altura_m',
    'weight_kg': 'peso_kg',
    'experience_years': 'anos_experiencia',
}

# Aplica somente renomeações de colunas existentes no export.
rename_applied = {}
for old, new in rename_cols_pt_nba.items():
    if old in df_export.columns:
        rename_applied[old] = new

# Renomeia colunas mantendo apenas o mapeamento válido.
df_export = df_export.rename(columns=rename_applied)

df_export.head()
```

## **Observações metodológicas**

- O pipeline **não remove linhas** por padrão, para evitar perda de
  informação sem justificativa analítica.
- As colunas originais são preservadas em `df` para rastreabilidade.
- Para exportação, usa-se `df_export` sem campos brutos não tratados:
  `salary`, `birthdate`, `height`, `weight` e `experience`.
- Para exportação, também são removidas todas as colunas com prefixo
  `avg`.
- Colunas sem dado (100% ausentes) são removidas de `df_export`.
- Antes da exportação, as colunas são renomeadas de EN -> PT-BR com
  termos técnicos da NBA.
- As medidas físicas derivadas seguem o **SI**: altura em metros
  (`altura_m`) e peso em quilogramas (`peso_kg`).
